#### Use plt.subplots(nrows=1, ncols=2) to create the plot.

```python
# Method 2
# Figure with 1 row and 2 columns of subplots
fig, ax = plt.subplots(nrows=1, ncols=2)

# (x, y) on the first subplot
ax[0].plot(x, y, color='orange', label='(x, y)')
ax[0].legend()

# (x, z) on the second subplot
ax[1].plot(x, z, color='blue', label='(x, z)')
ax[1].legend()

# Show the figure
plt.show()
```

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# your code here
# Resizing the plot and doing it all
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# (x, y) on the first subplot
ax[0].plot(x, y, color='orange', label='(x, y)')
ax[0].legend()
ax[0].set_title("(x,y)", fontweight='bold', fontsize=16)
ax[0].set_xlabel("X", fontsize=10)
ax[0].set_ylabel("Y", fontsize=10)

# (x, z) on the second subplot
ax[1].plot(x, z, color='blue', label='(x, z)')
ax[1].legend()
ax[1].set_title("(x,z)", fontweight='bold', fontsize=16)
ax[1].set_xlabel("X", fontsize=10)
ax[1].set_ylabel("Z", fontsize=10)

# Show the figure
plt.show()
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# your code here
# MEDIAN STEPS BY MONTH_ENCODED
fig, axs = plt.subplots(1, 2, figsize=(12, 4))
monthly_steps = fitbit.groupby(['Months_encoded'])['Steps'].median()
monthly_steps.plot(kind='bar', ax=axs[0], color = 'grey')
axs[0].set_title('Median Steps by Month')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Steps')

# MEDIAN STEPS BY WORK_OR_WEEKEND
weekly_steps = fitbit['Steps'].groupby(fitbit['Work_or_Weekend']).median()
plt.barh(weekly_steps.index, weekly_steps.values, height= 0.9, color= 'black', align='center', )
plt.yticks([0, 1], ['Weekend', 'Work'])
plt.show()

# Save the figure
plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

features = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']

for i, feature in enumerate(features):
    axs[i].scatter(fitbit[feature], fitbit['Steps'], color = 'purple')
    axs[i].set_xlabel(feature)
    axs[i].set_ylabel('Steps')
plt.show()
```
